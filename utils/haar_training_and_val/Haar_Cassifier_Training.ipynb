**Train Haar Cascade Classifier**

Check the OpenCV version

```
%%shell
pkg-config --modversion opencv
```

Get the dataset from the link

```
%%shell
wget 'https://www.dropbox.com/s/jyjp43fk3d2d0il/mask_detection_dataset.zip?dl=0' -O dataset.zip
unzip dataset.zip
cp -r mask_detection_dataset/* ./
```

Mount Google Drive to save the model

```
from google.colab import drive
drive.mount('/content/drive')
```

Create the VEC-file for the training

```
%%shell
opencv_createsamples -info info.dat -num 3432 -w 20 -h 20 -vec mask.vec
```

Train and save the model 

```
%%shell
mkdir output_model
opencv_traincascade -data output_model -vec mask.vec -bg bg.txt -numPos 2900 -numNeg 4129 -numStages 15 -w 20 -h 20 -bt DAB -precalcValBufSize 2048 -precalcIdxBufSize 2048 -mode ALL

mkdir 'drive/My Drive/Haar_log_model/log_15s'
cp output_model/* 'drive/My Drive/Haar_log_model/log_15s'

opencv_traincascade -data output_model -vec mask.vec -bg bg.txt -numPos 2900 -numNeg 4129 -numStages 20 -w 20 -h 20 -bt DAB -precalcValBufSize 2048 -precalcIdxBufSize 2048 -mode ALL

mkdir 'drive/My Drive/Haar_log_model/log_20s'
cp output_model/* 'drive/My Drive/Haar_log_model/log_20s'
```
